Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

Enter Hours: 45

Enter Rate: 10

Pay: 475.0

```python
def computepay(hours, rate):
    if hours > 40:
        regular_pay = 40 * rate
        overtime_hours = hours - 40
        overtime_pay = overtime_hours * rate * 1.5
        total_pay = regular_pay + overtime_pay
    else:
        total_pay = hours * rate
    return total_pay

hours_input = float(input("Enter Hours: "))
rate_input = float(input("Enter Rate: "))

pay = computepay(hours_input, rate_input)
print("Pay:", pay)
```

Challenge
Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

Score           Grade
[ 98, 100]      1.0

[ 94,  97)      1.25

[ 90,  93)      1.5

[ 86,  89)      1.75

[ 82,  85)      2.0

[ 78,  81)      2.25

[ 75,  77)      2.5

[ 0,   74)      5.0

Run the program repeatedly to test the various different values for input.

```python
def computegrade(score):
    if score < 0 or score > 100:
        return "Invalid score"
    elif score >= 98:
        return "1.0"
    elif score >= 94:
        return "1.25"
    elif score >= 90:
        return "1.5"
    elif score >= 86:
        return "1.75"
    elif score >= 82:
        return "2.0"
    elif score >= 78:
        return "2.25"
    elif score >= 75:
        return "2.5"
    else:
        return "5.0"

try:
    score_input = float(input("Enter score (0-100): "))
    grade = computegrade(score_input)
    print("Grade:", grade)
except ValueError:
    print("Invalid input. Please enter a numeric score.")
```
