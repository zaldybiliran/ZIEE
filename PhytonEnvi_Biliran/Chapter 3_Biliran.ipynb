EXERCISE 2

```python
hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

if hours > 40:
    regular_pay = 40 * rate
    overtime_hours = hours - 40
    overtime_pay = overtime_hours * rate * 1.5
    pay = regular_pay + overtime_pay
else:
    pay = hours * rate

print("Pay:", pay)
```

```python
score = float(input("Enter score (0-100): "))

if score < 0 or score > 100:
    print("Invalid score")
elif score >= 98:
    print("Grade: 1.0")
elif score >= 94:
    print("Grade: 1.25")
elif score >= 90:
    print("Grade: 1.5")
elif score >= 86:
    print("Grade: 1.75")
elif score >= 82:
    print("Grade: 2.0")
elif score >= 78:
    print("Grade: 2.25")
elif score >= 75:
    print("Grade: 2.5")
else:
    print("Grade: 5.0")
```
